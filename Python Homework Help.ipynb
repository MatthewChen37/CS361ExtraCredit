## Reading from CSVs

Given a CSV file, we can open it using Python's built in functions. We can even open text files as CSVs!
If you're curious about opening and reading files you can find more information here: https://www.w3schools.com/python/python_file_handling.asp


```python
import csv ## Get Python's CSV functions

reader = csv.reader(open("data/iris.data"), delimiter=",")
## Here, we open up the file using a file reader to get all of it's data
## Notice that the data is a .data file. We can actually open up several different types up files besides .csv files
## using the csv reader. A good tip is that if you can open it up in Notepad (on Windows) or TextEdit (on MacOS) or Vim (on Linux)
## and it looks human-interpretable similar to the iris.data file you can probably open it using a csv reader!
## Also even though csv stands for "comma-seperated values", you can change the delimiter as above 
## For example: csv.reader(open(file_name), delimiter= " ") reads space seperated values
## Another example: csv.reader(open(file_name), delimiter="\t") reads tab seperated values

```

### Iterating through the data

Try running this line of code AFTER running above (otherwise reader will not be defined).

```python
for line in reader:
    print(line)
```

Notice that each line in the reader is a list of all the data in the row of the data file.
To access specific data pieces in each list, we can use Python's list indexing functions. 
More information about it can be found here: https://www.w3schools.com/python/python_lists.asp

```python
reader = csv.reader(open("data/iris.data"), delimiter=",") # We have to reopen the reader because by iterating through all the lines it removes each line from the reader
for line in reader:
    print(line[0]) #Access 1st element of the list
    print(line[1]) #Access 2nd element of the list
    print(line[2]) #Access 3rd element of the list
    print(line[3]) #Access 4th element of the list
    print(line[4]) #Access 5th element of the list
    # Anything greater than 4 will result in an error 
```

## A quick introduction to MatPlotLib

Matplotlib will probably be one of the most useful libraries you'll use in your assignments because it allows you to plot charts, historgrams, etc. If you want to scour the documentation for more help, you can do so on their website: https://matplotlib.org/.

Let's first try plotting the iris data!

```python
import matplotlib.pyplot as plt #<-- We get the python plotting functions and when we call these functions we call it by using plt.FUNCTION_NAME

petal_length = [] #Define two lists in python and we will be comparing petal length (3rd element) and petal width (4th element)
petal_width = [] #We can't directly manipulate the line data because the features are sorted in columns in the original file

reader = csv.reader(open("data/iris.data"), delimiter=",") # We have to reopen the reader because by iterating through all the lines it removes each line from the reader
for line in reader:
    petal_length.append(float(line[2])) #putting third column into a list
    petal_width.append(float(line[3])) #putting fourth column into a list
    #make sure to cast it as a float (number with decimal points)
    #because the line reader will interpret it as a character/string!
    
    
print(petal_length)
print(petal_width)
```

After running the code above you'll see that we sorted the data into two lists. Now plotting it with matplotlib is as simple as below! Note that the first parameter will always represent that dependent variable/x-axis while any other datasets after will represent the independent/y-axis.

```python
plt.plot(petal_length, petal_width)
```

That does not seem like the correct plot we want! Without any parameters, the plot function assumes that we are plotting a continuous line and so will draw straight lines between each of the points. 

First let us look at making them points. By simplying adding the third parameter 'ro' after the two lists in plot, it speficies that we want red dots and that we don't want to draw lines between points.
    
    

```python
plt.plot(petal_length, petal_width, 'ro') 
```

We can make them blue squares by changing "r" to "b" and "o" to "s"

```python
plt.plot(petal_length, petal_width, 'bs') 
```

Now let's make the axes more readable. First let's look at a diagram of a plot:



If you ever have problems with the axes you can use these functions: (plt.xticks and plt.yticks)

```python
import numpy as np

x_ticks = np.arange(min(petal_length), max(petal_length) + 1.0, 0.5)
# we need to use numpy's arange function because we are dealing with floats, which python's range() function doesn't deal with
# don't worry if you don't know numpy, what this line simply does is make an array of numbers that starts from the lowest 
# petal_length and increments by 0.5 until ireaches the max petal length (does not include max(petal_length) +1 because its
# defined as non-inclusive)
plt.xticks(x_ticks)


plt.plot(petal_length, petal_width, 'ro')
```

Doing same thing with y axis:

```python
import numpy as np

y_ticks = np.arange(min(petal_width), max(petal_width) + 1.0, 0.5)

plt.yticks(y_ticks) #remember to feed the list into the plot!


plt.plot(petal_length, petal_width, 'ro')
```

Now let's add titles to the x axis, y axis and to the whole graph! It's as simple as this:

```python
plt.xlabel("petal length")
plt.ylabel("petal width")
plt.title("petal length vs. petal width")
plt.plot(petal_length, petal_width, 'ro')
```

Another cool thing you can do is to plot individual points and add labels to them (you might find this useful somewhere :O )

```python
plt.xlabel("petal length")
plt.ylabel("petal width")
plt.title("petal length vs. petal width")
for i in range(len(petal_width)):
    plt.plot(petal_length[i], petal_width[i], 'ro', markersize = 10)
    plt.text(petal_length[i] * (1 + 0.01), petal_width[i] * (1 + 0.01) , "M", fontsize=12)

    
```

Let's also plot a line of best fit which you may find useful somewhere :)

```python
print(np.corrcoef(petal_length, petal_width)) # think of this as a 2d array 
coeff = np.corrcoef(petal_length, petal_width)[0][1]
print(coeff) # This is how you will get the correlation coefficient of two lists! (Note: must first import numpy as np)
```

```python
PLmean = np.mean(petal_length) # Just throw your lists into these numpy functions and they'll do the rest!
PLstd = np.std(petal_length)

PWmean = np.mean(petal_width)
PWstd = np.std(petal_width)

print(PLmean)
print(PLstd)
print(PWmean)
print(PWstd)
```

Now let's plot the line:

```python
plt.xlabel("petal length")
plt.ylabel("petal width")
plt.title("petal length vs. petal width")
plt.plot(petal_length, petal_width, 'ro')

x = np.linspace(min(petal_length), max(petal_length), 1000)
#similar to np.arange generates a list but includes last number
#and instead of defining the step we define that there are 1000 intervals between the min and max

plt.plot(x, PWstd * coeff * ((x - PLmean) / PLstd) + PWmean, '--k')


```
